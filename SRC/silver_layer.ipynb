```python
from pyspark.sql.functions import *
from pyspark.sql.window import *
from pyspark.sql.types import *
```

```python
df=spark.read.format('delta').option('inferSchema',True).option('header','True').load('/Volumes/dog_warehouse/default/dog_data/Bronze/Raw_Data')
```

```python
df.display()
```

```python
print(df.count())
df=df.dropDuplicates(subset=['id','attributes_name'])
print(df.count())
```

```python
hypo_allergenic_dogs=df.filter(col('attributes_hypoallergenic')=='true')
non_allergenic_dogs=df.filter(col('attributes_hypoallergenic')=='false')
```

```python
hypo_allergenic_dogs=hypo_allergenic_dogs.drop('attributes_hypoallergenic').drop('id')
hypo_allergenic_dogs.display()
```

```python
non_allergenic_dogs=non_allergenic_dogs.drop('attributes_hypoallergenic').drop('id')
non_allergenic_dogs.display()
```

```python
hypo_allergenic_dogs.write.format('delta').option('writeOptimze',True).mode('overwrite').save('/Volumes/dog_warehouse/default/dog_data/Silver/Allergic')
non_allergenic_dogs.write.format('delta').option('writeOptimze',True).mode('overwrite').save('/Volumes/dog_warehouse/default/dog_data/Silver/Non_Allergic')
```
