```python

Bronze_Path='/Volumes/dog_warehouse/default/dog_data/Bronze'
Silver_Path='/Volumes/dog_warehouse/default/dog_data/Silver'
Archive_Path='/Volumes/dog_warehouse/default/dog_data/Archive'
```

```python
import os,shutil,datetime

suffix=datetime.datetime.now().strftime('%Y%m%d')

def archive_util(SRC,DST):
    for item in os.listdir(SRC):
        print(item)
        if os.path.isdir(os.path.join(SRC,item)):
            dst_folder = os.path.join(DST, f"{item}_{suffix}")
            src_folder = os.path.join(SRC, item)
            #shutil.make_archive(dst_folder+'_archived', 'zip', src_folder)
            shutil.move(src_folder, dst_folder)
            print('Archived',dst_folder+'_archived')
            





```

```python
archive_util(Bronze_Path, Archive_Path)

```

```python
archive_util(Silver_Path,Archive_Path)
```
